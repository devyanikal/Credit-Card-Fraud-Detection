

---


**Taking 492 random transactions/rows from legit transactions and joining it with 492 fraud transactions/rows so that model can learn to clearly distinguish between the two classes for new data.**

---



```python
legit_trans2=legit_trans.sample(n=492)
```

```python
legit_trans2.shape
```

```python
#joining legit_trans2 and fraud_trans dataframes
credit_card_data2=pd.concat([legit_trans2,fraud_trans],axis=0)
```

```python
credit_card_data2.shape
```

```python
credit_card_data2.head(5)
```

```python
#No. of legits and fraud transactions
credit_card_data2['Class'].value_counts()
```

```python
#comparing mean values of both the classes
credit_card_data2.groupby('Class').mean()
```

Therefore we can say that nature of the dataset is not changed.

Separating features from the target variable



```python
X=credit_card_data2.drop(columns='Class',axis=1)
```

```python
Y=credit_card_data2['Class']
```

```python
X
```

```python
Y
```

Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=1)
```

```python
print("Feature data size: ",X.shape)
print("Training data size: ",X_train.shape)
print("Testing data size: ",X_test.shape)
```

Using Logistic Regression for binary classification.

```python
from sklearn.linear_model import LogisticRegression
```

TRAINING MODEL WITH TRAINING DATA

```python
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#Finding Accuracy score on training data
X_train_predict=model.predict(X_train)
trainingdata_accuracy_score=accuracy_score(X_train_predict,Y_train)
```

```python
print('Accuracy score of training data is ',trainingdata_accuracy_score)
```

```python
print('Accuracy percentage of training data: ',trainingdata_accuracy_score*100,'%')
```

```python
#Finding Accuracy score on test data
X_test_predict=model.predict(X_test)
testdata_accuracy_score=accuracy_score(X_test_predict,Y_test)
```

```python
print('Accuracy score of test data is ',testdata_accuracy_score)
```

```python
print('Accuracy percentage of test data: ',testdata_accuracy_score*100,'%')
```
